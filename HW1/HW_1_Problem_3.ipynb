### Imports

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

# Problem 3

### Importing and Loading MNIST

```
import tensorflow as tf

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

### Function for getting all images of specified number

```
def getAllOf(number):
  filter = np.where(y_train == number)
  return x_train[filter]
```

### Function for getting average of each pixel for all of a certain number

```
def getAvgOf(number):
  allNums = getAllOf(number)
  return np.average(allNums, axis=0)
```

### Function to display averages of specified number

```
def displayAvgOf(number):
  avg = getAvgOf(number)
  plt.imshow(avg, interpolation='nearest', cmap=cm.gray)
  plt.show()
```

## Display Averages of Each Number

```
for i in range(10):
  displayAvgOf(i)
```
