### Imports

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

# Problem 1

### Create Meshgrid

```
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
```

### Plotting Function

```
def plotFunc(func):
  Z = func(X, Y)
  
  fig = plt.figure(figsize=(20, 10))
  ax = fig.gca(projection='3d')
  
  surf = ax.plot_surface(X, Y, Z, cmap=cm.spring, linewidth=0, antialiased=False, zorder=-1)
  
  fig.colorbar(surf, shrink=0.5, aspect=5)
  
  return fig, ax
```

## Function 1

### Create data for plot 1

```
@np.vectorize
def func1(x, y):
  return x**2 + y**2
```

### Plot Function 1

```
fig, ax = plotFunc(func1)

# Mark Minima
ax.scatter(0, 0, 0, marker='o', s=1000, c='k')
ax.text(1.5, -2, -1, '(Global) Minima', size=25)

plt.show()
```

## Function 2

### Create data for plot 2

```
@np.vectorize
def func2(x, y):
  return -(x**2 + y**2)
```

### Plot function 2

```
fig, ax = plotFunc(func2)

# Mark Maxima
ax.scatter(1, -2, 10, marker='o', s=1000, c='k')
ax.text(0, 2, 2, '(Global) Maxima', size=25)

plt.show()
```

## Function 3

### Create data for plot 3

```
@np.vectorize
def func3(x, y):
  return (x**3 + y**3)
```

### Plot function 3

```
fig, ax = plotFunc(func3)

# Mark saddle point
ax.autoscale(False)
ax.scatter(5, -10, 270, marker='o', s=1000, c='k', zorder=1)
ax.text(0, 0, 40, 'Saddle Point', size=25)

plt.show()
```
